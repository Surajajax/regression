```python
x=data.squareMeters
y=data.price
table= pd.DataFrame({'squareMeters': x, 'price': y})
print(table)
```

```python
x_array=np.array(x)
y_array=np.array(y)
print(x_array)
print(y_array)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_array, y_array, marker='.',c='r')
plt.title("Housing Price")
plt.xlabel("SquareMeter")
plt.ylabel("Price")
plt.show()
```

```python
def cost_model(x,y ,w, b):
    n=x_array.shape[0]
    pred_y=np.zeros(n)
    cost = 0
    
    
    for i in range(n):
        pred_y= w * x[i] + b
        cost = cost + (pred_y - y[i])**2
    total_cost = 1 / (2 * n) * cost
        
    return total_cost
```

```python
def compute_gradient(x, y, w, b): 
    n = x_array.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(n):  
        pred_y = w * x[i] + b 
        dj_dw_i = (pred_y- y[i]) * x[i] 
        dj_db_i = pred_y - y[i] 
        dj_db += dj_db_i
        dj_dw += dj_dw_i 
    dj_dwe = dj_dw / n 
    dj_dbe = dj_db / n 
        
    return dj_dwe, dj_dbe
```

```python
import math
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, gradient_function): 

    J_history = []
    p_history = []
    b = b_in
    w = w_in
    
    for i in range(num_iters):
        
        dj_dwe, dj_dbe = gradient_function(x, y, w , b)     

        
        b = b - alpha * dj_dbe                            
        w = w - alpha * dj_dwe                            

        
        if i<100000:      
            J_history.append( cost_function(x, y, w , b))
            p_history.append([w,b])
        
        if i% math.ceil(num_iters/10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                f"dj_dw: {float(dj_dwe): 0.3e}, dj_db: {float(dj_dbe): 0.3e}  ",
                f"w: {w: 0.3e}, b: {b: 0.5e}")


 
    return w, b, J_history, p_history 

```

```python
w_init = 0
b_init = 0

iterations = 10000
tmp_alpha = 0.01


x_scaled = (x_array - x_array.mean()) / x_array.std()
y_scaled = (y_array - y_array.mean()) / y_array.std()


w_final, b_final, J_hist, p_hist = gradient_descent(x_scaled,y_scaled, w_init, b_init, tmp_alpha, 
                                                    iterations, cost_model, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

```python
plt.scatter(x_scaled, y_scaled, marker='.',c='r')
plt.title("Housing Price")
plt.xlabel("SquareMeter")
plt.ylabel("Price")
plt.show()
```

```python
print(f"square:{w_final*75523+b_final:0.1f}")
```

```python
mu_x = np.mean(x_array, axis=0)
std_x = np.std(x_array, axis=0)

mu_y = np.mean(y_array)
std_y = np.std(y_array)
```

```python
w_actual = (w_final / std_x) * std_y
b_actual = std_y * (b_final - np.sum((w_final* mu_x) / std_x)) + mu_y
```

```python
def t_model(x, y, w, b):
    n = x.shape[0] 
    y_pred = np.zeros(n)

    for i in range(n):
        y_pred[i] = w * x[i] + b

    return y_pred
```

```python
s=t_model(x_array,y_array,w_actual,b_actual)
print(s)
```
