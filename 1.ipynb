```python
x=data.squareMeters
y=data.price
table= pd.DataFrame({'squareMeters': x, 'price': y})
print(table)
```

```python
x_array=np.array(x)
y_array=np.array(y)
print(x_array)
print(y_array)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_array, y_array, marker='.',c='r')
plt.title("Housing Price")
plt.xlabel("SquareMeter")
plt.ylabel("Price")
plt.show()
```

```python
def cost_model(x,y ,w, b):
    n=x_array.shape[0]
    pred_y=np.zeros(n)
    cost = 0
    
    
    for i in range(n):
        pred_y= w * x[i] + b
        cost = cost + (pred_y - y[i])**2
    total_cost = 1 / (2 * n) * cost
        
    return total_cost
```

```python
def compute_gradient(x, y, w, b): 
    n = x_array.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(n):  
        pred_y = w * x[i] + b 
        dj_dw_i = (pred_y- y[i]) * x[i] 
        dj_db_i = pred_y - y[i] 
        dj_db += dj_db_i
        dj_dw += dj_dw_i 
    dj_dwe = dj_dw / n 
    dj_dbe = dj_db / n 
        
    return dj_dwe, dj_dbe
```

```python
import math
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, gradient_function): 

    
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    p_history = []
    b = b_in
    w = w_in
    
    for i in range(num_iters):
        # Calculate the gradient and update the parameters using gradient_function
        dj_dwe, dj_dbe = gradient_function(x, y, w , b)     

        # Update Parameters using equation (3) above
        b = b - alpha * dj_dbe                            
        w = w - alpha * dj_dwe                            

        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( cost_function(x, y, w , b))
            p_history.append([w,b])
        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters/10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                f"dj_dw: {float(dj_dwe): 0.3e}, dj_db: {float(dj_dbe): 0.3e}  ",
                f"w: {w: 0.3e}, b: {b: 0.5e}")


 
    return w, b, J_history, p_history #return w and J,w history for graphing

```

```python
# initialize parameters
w_init = 0
b_init = 0
# some gradient descent settings
iterations = 10000
tmp_alpha = 0.01

# Feature scaling (min-max or z-score)
x_scaled = (x_array - x_array.mean()) / x_array.std()
y_scaled = (y_array - y_array.mean()) / y_array.std()

# run gradient descent
w_final, b_final, J_hist, p_hist = gradient_descent(x_scaled,y_scaled, w_init, b_init, tmp_alpha, 
                                                    iterations, cost_model, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```
